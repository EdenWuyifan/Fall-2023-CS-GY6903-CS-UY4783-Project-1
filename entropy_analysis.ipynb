```python
from enc import ctoi
from collections import Counter
import math

import matplotlib.pyplot as plt
```

```python
def diff(i_c, i_p):
    d = i_c - i_p
    if d < 0:
        d = d + 27
    return d

def entropy(counter: Counter):
    ent = 0.
    for cnt in counter.values():
        p = (cnt + 1) / sum(counter.values())
        ent -= p * math.log(p)
    return ent
```

```python
def text_diff(ciphertext, plaintext, silent=True):
    plain_stream = [ctoi[p] for p in plaintext]
    cipher_stream = [ctoi[c] for c in ciphertext]
    diffs = [diff( i_c, i_p ) for i_c, i_p in zip(cipher_stream, plain_stream)]

    if not silent:
        print(diffs)

    return Counter(diffs)
```

```python
def progressing_entropy_analysis(ciphertext, plaintext, end=100, keylen=4):
    X, Y = [], []
    for step in range(keylen, end, keylen):
        d = text_diff(ciphertext[:step], plaintext[:step])
        ent = entropy(d)
        X.append(step)
        Y.append(ent)

    return X, Y
```

```python
fig, ax_matrix = plt.subplots(5, 1, figsize=(12, 18))

for i in range(5):
    for j in range(5):
        X, Y = progressing_entropy_analysis(ciphers[i], plains[j], end=200)
        ax_matrix[i].plot(X, Y, label=f'plain {j+1}')
        ax_matrix[i].legend()
        ax_matrix[i].set_title(f'cipher {i+1}')

fig.tight_layout()
```
